## Downloading and parsing the CIFAR-10 dataset

Download the [CIFAR-10](http://www.cs.toronto.edu/~kriz/cifar.html) dataset for Python. The file is 163 MB so executing the following cell would take a while.

```python
import urllib.request

urllib.request.urlretrieve("http://www.cs.toronto.edu/~kriz/cifar-10-python.tar.gz", "cifar-10-python.tar.gz")
```

Check that md5sum is equal to `c58f30108f718f92721af3b95e74349a`

```python
import hashlib

def md5(fname):
    hash_md5 = hashlib.md5()
    with open(fname, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()

print(md5('cifar-10-python.tar.gz'))
```

Unpickle the data

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict

dicts = []
for i in range(5):
    dicts.append(unpickle('cifar-10-batches-py/data_batch_' + str(i + 1)))
```

Combine all batches into a numpy array of shape (50000, 3072)

```python
import numpy as np

data = np.zeros([50000, 3072])
for i in range(5):
    data[i * 10000:(i + 1) * 10000] = dicts[i][b'data']
data.shape
```

Display an image to make sure everything is working

```python
import matplotlib.pylab as plt

single_image = np.transpose(np.reshape(data[15], (3, 32, 32)), (1, 2, 0))
plt.imshow(single_image.astype(np.uint8))
```

Choose randomly 25000 images to embed messages in them. Save all images to files.

```python
import matplotlib
import random
import os

data = data.astype(np.uint8)

random.shuffle(data)

data_fishy = data[:25000]
data_clean = data[25000:]

newpaths = [r'fishy', 'clean']
for newpath in newpaths:
    if not os.path.exists(newpath):
        os.makedirs(newpath)

fishy_paths = []

for i in range(data_fishy.shape[0]):
    matplotlib.image.imsave(f"fishy/image{i + 1}.png", \
                            np.transpose(np.reshape(data[i], (3, 32, 32)), (1, 2, 0)))
    fishy_paths.append(f"fishy/image{i + 1}.png")
    
for i in range(data_clean.shape[0]):
    matplotlib.image.imsave(f"clean/image{i + 1}.png", \
                            np.transpose(np.reshape(data[i], (3, 32, 32)), (1, 2, 0)))
```

A function that generates random strings with a given length

```python
import random
import string

def generate_strings(n, length):
    result = []
    for i in range(n):
        result.append(''.join(random.choices(string.ascii_uppercase + string.digits, k=10)))
    return result
```

Generate 25000 random 10-character strings and save them to a file

```python
f = open('strings.txt', 'w')

l = generate_strings(25000, 10)

for item in l:
    f.write(f"{item}\n")
```

Save fishy file paths to two seperate files

```python
f1 = open('imagein.txt', 'w')
f2 = open('imageout.txt', 'w')

for item in fishy_paths:
    f1.write(f"{item}\n")
    f2.write(f"{item}\n")
```

Read fishy files

```python
from os import listdir
from os.path import isfile, join

import cv2

fishy_image_files = [join('fishy', f) for f in listdir('fishy') if isfile(join('fishy', f))]
clean_image_files = [join('clean', f) for f in listdir('clean') if isfile(join('clean', f))]

data_fishy = np.zeros([len(fishy_image_files), 3072])
data_clean = np.zeros([len(clean_image_files), 3072])

for i, image_file in enumerate(fishy_image_files):
    data_fishy[i] = cv2.imread(image_file).flatten()
    
for i, image_file in enumerate(clean_image_files):
    data_clean[i] = cv2.imread(image_file).flatten()
```

```python
data_fishy = np.hstack([data_fishy, np.ones([25000, 1])])
data_clean = np.hstack([data_clean, np.zeros([25000, 1])])
data = np.vstack([data_fishy, data_clean])
```

```python
import pandas as pd

data = pd.DataFrame(data)

data.to_csv('data.csv', index=None)
```

```python
import pandas as pd

data = pd.read_csv('data.csv')

train, validate, test = np.split(data.sample(frac=1, random_state=55), [int(.6 * len(data)), int(.8 * len(data))])
```

```python
X_train, y_train = train.values[:, :-1], train.values[:, -1]
X_validate, y_validate = validate.values[:, :-1], validate.values[:, -1]
X_test, y_test = test.values[:, :-1], test.values[:, -1]
```

```python
X_train = X_train / 255
X_validate = X_validate / 255
X_test = X_test / 255
```

```python
from keras.models import Sequential
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
from keras.layers import Input, Dense, Dropout, Flatten, \
Conv2D, MaxPooling2D, Reshape, BatchNormalization

model = Sequential()
model.add(Reshape((32, 32, 3), input_shape=(3072, )))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=Adam(lr=0.001),
              metrics=['accuracy'])

model.fit(X_train, y_train,
          batch_size=512,
          epochs=100,
          verbose=1,
          validation_data=(X_validate, y_validate),
          callbacks=[EarlyStopping(patience=2)])
```
